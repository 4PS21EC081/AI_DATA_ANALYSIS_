```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load a sample dataset (you can replace this with your own DataFrame)
iris_df = sns.load_dataset('iris')

# Calculate the correlation matrix
correlation_matrix = iris_df.corr(numeric_only=True)

# --- Advanced Styling of the Heatmap ---

# 1. Set figure size
plt.figure(figsize=(8, 7))

# 2. Create the heatmap with advanced styling
sns.heatmap(correlation_matrix,
            annot=True,             # Display correlation values in the cells
            cmap='coolwarm',        # Use a diverging color map
            fmt=".2f",              # Format the annotations to two decimal places
            linewidths=.5,         # Add lines between cells
            cbar_kws={'shrink': .8}, # Adjust the size of the color bar
            square=True,            # Ensure the heatmap is square
            mask=None)              # You can use a mask to hide parts of the heatmap

# 3. Customize labels and title
plt.title('Correlation Matrix of Iris Features', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)

# 4. Adjust color bar label (optional)
cbar = plt.gca().collections[0].colorbar
cbar.set_label('Correlation Coefficient', fontsize=12)

# 5. Improve layout
plt.tight_layout()

# Display the heatmap
plt.show()

# --- Optional: Using a Mask to Hide the Upper Triangle ---
plt.figure(figsize=(8, 7))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            linewidths=.5,
            cbar_kws={'shrink': .8},
            square=True,
            mask=mask)
plt.title('Correlation Matrix of Iris Features (Upper Triangle Masked)', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.tight_layout()
plt.show()
```
