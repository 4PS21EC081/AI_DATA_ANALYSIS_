```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset from Seaborn (you can replace this with your own DataFrame)
iris_df = sns.load_dataset('iris')

# Display the first few rows to understand the data
print("First 5 rows of the Iris DataFrame:")
print(iris_df.head())
print("\n")

# Create the pairplot using Seaborn
plt.figure(figsize=(10, 10)) # Optional: Adjust figure size
sns.pairplot(iris_df)
plt.suptitle('Pairwise Relationships in the Iris Dataset (Seaborn)', y=1.02) # Add a title to the entire figure
plt.show()

# --- Optional: Include a hue for categorical variables ---
# You can color-code the scatter plots and distributions based on a categorical variable
plt.figure(figsize=(10, 10))
sns.pairplot(iris_df, hue='species')
plt.suptitle('Pairwise Relationships in the Iris Dataset by Species (Seaborn)', y=1.02)
plt.show()

# --- Optional: Customize the diagonal plots ---
# You can specify the type of plot to show on the diagonal (e.g., histograms or KDEs)
plt.figure(figsize=(10, 10))
sns.pairplot(iris_df, diag_kind='kde', hue='species')
plt.suptitle('Pairwise Relationships with KDE on Diagonal by Species (Seaborn)', y=1.02)
plt.show()
```
