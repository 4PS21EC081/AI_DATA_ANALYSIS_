```python
import matplotlib.pyplot as plt

# Data for the bar plot
products = ['A', 'B', 'C', 'D', 'E']
sales_in_thousands = [23, 75, 57, 90, 65]

# Create the bar plot
plt.figure(figsize=(8, 6))  # Optional: Adjust figure size for better readability
plt.bar(products, sales_in_thousands)

# Add labels and title
plt.xlabel('Products')
plt.ylabel('Sales (in thousands)')
plt.title('Number of Sales for Each Product')

# Add the exact sales figures on top of the bars
for product, sale in zip(products, sales_in_thousands):
    plt.text(product, sale, str(sale), ha='center', va='bottom')

# Display the plot
plt.show()
```
