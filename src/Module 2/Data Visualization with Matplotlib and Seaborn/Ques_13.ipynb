```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris_df = sns.load_dataset('iris')

# --- Customizing the Boxplot ---

# 1. Set a different color palette
sns.set_palette("viridis")  # Try other palettes like "Set2", "pastel"

# 2. Create the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_length', data=iris_df, linewidth=1.5, fliersize=4)

# 3. Customize labels and title
plt.xlabel('Species', fontsize=14)
plt.ylabel('Petal Length (cm)', fontsize=14)
plt.title('Distribution of Petal Length by Species (Seaborn)', fontsize=16, fontweight='bold')

# 4. Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 5. Customize tick parameters
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Remove top and right spines for a cleaner look (optional)
sns.despine()

# Display the plot
plt.tight_layout()
plt.show()

# --- Optional Customizations ---

# You can further customize individual box properties if needed
# For example, changing the color of the boxes:
# box_colors = ['lightcoral', 'lightseagreen', 'lightskyblue']
# plt.figure(figsize=(8, 6))
# sns.boxplot(x='species', y='petal_length', data=iris_df, palette=box_colors)
# plt.xlabel('Species', fontsize=14)
# plt.ylabel('Petal Length (cm)', fontsize=14)
# plt.title('Distribution of Petal Length by Species (Custom Colors)', fontsize=16, fontweight='bold')
# plt.tight_layout()
# plt.show()
```
